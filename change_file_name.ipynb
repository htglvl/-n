```python
import os

# Specify the directory containing the files
directory = '..\\raw_data\\'

# Set the starting index for renaming
start_index = 1358

# Get a list of all files in the directory that match the naming pattern
files = sorted([f for f in os.listdir(directory) if f.startswith('dm_test_auto_') and f.endswith('_Tuan.pkl')])

# Rename the files with the new index
for i, file in enumerate(files):
    new_index = start_index + i
    new_name = f'dm_test_auto_{new_index}.pkl'
    old_file_path = os.path.join(directory, file)
    new_file_path = os.path.join(directory, new_name)
    
    os.rename(old_file_path, new_file_path)
    print(f'Renamed: {file} -> {new_name}')
```
